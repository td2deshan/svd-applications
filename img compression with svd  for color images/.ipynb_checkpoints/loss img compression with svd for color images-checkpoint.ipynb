```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import randomized_svd
from fbpca import pca

%matplotlib inline
```

## loss color img compression

```python
# load image in gray scale color & display

#opencv(cv2) color img in "BGR" format
img = cv2.imread('t.jpg')

plt.figure(figsize=(9, 6))

#For convert into "RGB" format
#b, g, r = cv2.split(img)
#img = cv2.merge((r,g,b))

plt.imshow(img, cmap='gray')
```

```python
#after applying svd algorithm img values not in 0-255 or 0.-1. range
#therefore have to scale
#here scale into 0.-1. range
def scale_img(x, minVal, maxVal):
    return (x - minVal)/(maxVal - minVal)
    
scale_img = np.vectorize(scale_img)
```

```python
def compress_img(img, algorithm, rank=50, bgr=True):
    if bgr:
        b, g, r = cv2.split(img)
        channels = [r, g, b]
    else:
        channels = list(cv2.split(img))
   
    for i in range(3):
        #apply reduced rank svd
        U, D, V = algorithm(channels[i], rank)
        #recreating img
        channels[i] = U @ np.diag(D) @ V
        #scaling img
        channels[i] = scale_img(channels[i], channels[i].min(), channels[i].max())
    #merging each channels into "RGB" & return
    return cv2.merge((channels[0], channels[1], channels[2]))
    
```

```python
#with sklearn --> randomized_svd
ret_img =compress_img(img, randomized_svd)
plt.imshow(ret_img)
```

```python
#with fbpca --> pca
ret_img =compress_img(img, pca)
plt.imshow(ret_img)
```

```python
import time
```

```python
def loop_unroll(img, algorithm, rank=50, bgr=True):
    if bgr:
        b, g, r = cv2.split(img)
    else:
        r, g, b = list(cv2.split(img))
   
    Ur, Dr, Vr = algorithm(r, rank)
    Ug, Dg, Vg = algorithm(g, rank)
    Ub, Db, Vb = algorithm(b, rank)
    
    r = Ur @ np.diag(Dr) @ Vr
    g = Ug @ np.diag(Dg) @ Vg
    b = Ub @ np.diag(Db) @ Vb
    
    r = scale_img(r, r.min(), r.max())
    g = scale_img(g, g.min(), g.max())
    b = scale_img(b, b.min(), b.max())
    
    return cv2.merge((r, g, b))
```

```python
t = time.time()
ret_img =compress_img(img, randomized_svd, rank=200)
print('time spend :', time.time()-t)
plt.imshow(ret_img)
```

```python
t = time.time()
ret_img = loop_unroll(img, randomized_svd, rank=200)
print('time spend :', time.time()-t)
plt.imshow(ret_img)
```
